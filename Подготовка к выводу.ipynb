```python
GAZP_model_direction = joblib.load("/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Models/trend_classification_model_GAZP.pkl")
GAZP_model_strength = joblib.load("/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Models/strength_classification_model_GAZP.pkl")
GAZP_scaler = joblib.load('/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Normalizer/strength_classification_normalizer_GAZP.pkl')
```

```python
symbols = ['GAZP', 'LKOH', 'NVTK', 'ROSN', 'SIBN', 'SNGS']
```

```python
result = ""

for symbol in symbols:
  model_direction = joblib.load(f"/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Models/trend_classification_model_{symbol}.pkl")
  model_strength = joblib.load(f"/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Models/strength_classification_model_{symbol}.pkl")
  scaler = joblib.load(f'/content/drive/MyDrive/MTS.Project/Strategy/Trend_classification/Normalizer/strength_classification_normalizer_{symbol}.pkl')
  
  data = pd.read_csv(f'/content/drive/MyDrive/MTS.Project/DataSet/GazNeft_D1/{symbol}_D1_HLOCV_MTS_Project_4_8_2021___5-41-42.csv', sep = ';')[['TIMEDATE','Close']].tail(11)
  data['TIMEDATE'] = pd.to_datetime(data['TIMEDATE'])
  df = data.diff()[1:]['Close']
  X = df
  direction = model_direction.predict([X])
  strength = model_strength.predict([X])

    
  
def get_recomend_trend_strat(symbol):
  if direction == 1:
    # Направление вверх
    text_rec_direct = 'Предположительное направление тренда – вверх'
  else:
    # Направление вниз
    text_rec_direct = 'Предположительное направление тренда – вниз'


  if strength == 1:
    # Слабый тренд
    text_rec_strength = 'Вероятность дальнейшего продолжения тренда – средняя'
  elif strength == 2:
    # Сильный тренд
    text_rec_strength = 'Вероятность дальнейшего продолжения тренда – большая'
  else:
    # Отсутствие тренда
    text_rec_strength = 'Вероятность дальнейшего продолжения тренда – маленькая'

  if strength == 2:
    if direction == 0:
      result += f'\n\t{symbol}: Продавать ({text_rec_direct}; {text_rec_strength})'
    else:
      result += f'\n\t{symbol}: Покупать ({text_rec_direct}; {text_rec_strength})'

  if result == "":
    result_text = '\n\tРекомендаций нет'
  else:
    result_text = result
```
